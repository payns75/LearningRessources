## ðŸ“˜ 1. Creating a List

```python
empty_list = []
print(empty_list)

todo_list = ['Learn Python List', 'How to manage List elements']
print(todo_list)

numbers = [1, 3, 2, 7, 9, 4]
print(numbers)

coordinates = [[0, 0], [100, 100], [200, 200]]
print(coordinates)
```

## ðŸŽ¯ 2. Accessing and Modifying Elements

```python
numbers = [1, 3, 2, 7, 9, 4]
print(numbers[0])  # First element
print(numbers[-1]) # Last element
print(numbers[-2]) # Second-to-last element

# Modify values
numbers[0] = 10
print(numbers)

numbers[2] /= 2
print(numbers)
```

## âž• 3. Adding and Removing Elements

```python
numbers = [1, 3, 2, 7, 9, 4]

numbers.append(100)            # Add to the end
numbers.insert(2, 100)         # Insert at position
print(numbers)

del numbers[0]                 # Delete by index
last = numbers.pop()           # Remove and return last element
print(last, numbers)

numbers.remove(9)              # Remove by value
print(numbers)
```

## ðŸ”¢ 4. Sorting a List

```python
guests = ['James', 'Mary', 'John', 'Patricia', 'Robert', 'Jennifer']

# Standard sort
print(sorted(guests))

# Reverse (in-place)
guests.sort(reverse=True)
print(guests)
```

### Sorting by a Specific Key

```python
companies = [
    ('Google', 2019, 134.81),
    ('Apple', 2019, 260.2),
    ('Facebook', 2019, 70.7)
]

def sort_key(company):
    return company[2]

companies.sort(key=sort_key, reverse=True)
print(companies)

# Simplified with lambda
companies.sort(key=lambda c: c[2])
print(companies)
```

## ðŸ§® 5. Creating a New Sorted List

```python
guests = ['James', 'Mary', 'John', 'Patricia', 'Robert', 'Jennifer']
sorted_guests = sorted(guests, reverse=True)
print(sorted_guests)

# Original remains unchanged
companies = [
    ('Google', 2019, 134.81),
    ('Apple', 2019, 260.2),
    ('Facebook', 2019, 70.7)
]

result = sorted(companies, key=lambda c: c[2], reverse=True)
print(result)
```

## âœ‚ï¸ 6. List Slicing

> sub_list = list[begin: end: step]

```python
colors = ['red', 'orange', 'yellow', 'green', 'blue', 'indigo', 'violet']

print(colors[1:4])     # From 2nd to 4th
print(colors[:3])      # First 3 elements
print(colors[-3:])     # Last 3 elements
print(colors[::2])     # Every other element
print(colors[::-1])    # Reverse

# Substitution
colors[0:2] = ['black', 'white']
print(colors)

colors[0:2] = ['black', 'white', 'gray']
print(colors)

del colors[2:5]        # Partial deletion
print(colors)
```

## ðŸ“¦ 7. List Unpacking

```python
colors = ['red', 'blue', 'green']
red, blue, green = colors

colors = ['cyan', 'magenta', 'yellow', 'black']
cyan, magenta, *other = colors
print(cyan, magenta, other)
```

## ðŸ” 8. Looping Through a List

```python
cities = ['New York', 'Beijing', 'Cairo', 'Mumbai', 'Mexico']

# Basic loop
for city in cities:
    print(city)

# With indexor index, city in enumerate(cities):
    print(f"{index}: {city}")

# Start index manually
for index, city in enumerate(cities, 0):
    print(f"{index}: {city}")
```

## ðŸ” 9. Finding an Element

```python
cities = ['New York', 'Beijing', 'Cairo', 'Mumbai', 'Mexico']

result = cities.index('Mumbai')
print(result)

city = 'Osaka'
if city in cities:
    print(f"{city} found at index {cities.index(city)}")
else:
    print(f"{city} not in the list")
```

## âš™ï¸ 10. Transforming Lists with map()

> iterator = map(fn, list)

```python
bonuses = [100, 200, 300]
iterator = map(lambda b: b * 2, bonuses)
print(list(iterator))

names = ['david', 'peter', 'jenifer']
print(list(map(lambda n: n.capitalize(), names)))

carts = [['SmartPhone', 400], ['Tablet', 450], ['Laptop', 700]]
TAX = 0.1
carts = map(lambda item: [item[0], item[1], item[1] * TAX], carts)
print(list(carts))
```

## ðŸ”Ž 11. Filtering Lists with filter()

> filter(fn, list)

```python
scores = [70, 60, 80, 90, 50]
filtered = filter(lambda s: s >= 70, scores)
print(list(filtered))

countries = [
    ['China', 1394015977],
    ['United States', 329877505],
    ['India', 1326093247],
    ['Indonesia', 267026366],
    ['Bangladesh', 162650853],
    ['Pakistan', 233500636],
    ['Nigeria', 214028302],
    ['Brazil', 21171597],
    ['Russia', 141722205],
    ['Mexico', 128649565]
]

populated = filter(lambda c: c[1] > 300000000, countries)
print(list(populated))
```

## ðŸ§® 12. Reducing Lists with reduce()


```python
from functools import reduce

scores = [75, 65, 80, 95, 50]
total = reduce(lambda a, b: a + b, scores)
print(total)
```

## ðŸ’¡ 13. List Comprehensions

```python
numbers = [1, 2, 3, 4, 5]
squares = [n**2 for n in numbers]
print(squares)

# Equivalent to:
squares = list(map(lambda n: n**2, numbers))
print(squares)

mountains = [
    ['Makalu', 8485],
    ['Lhotse', 8516],
    ['Kanchendzonga', 8586],
    ['K2', 8611],
    ['Everest', 8848]
]

highest = [m for m in mountains if m[1] > 8600]
print(highest)
```
